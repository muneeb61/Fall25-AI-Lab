```python
from collections import deque

class Graph:
    def __init__(self, adjacency_list):
        self.adjacency_list = adjacency_list

    def get_neighbors(self, v):
        return self.adjacency_list[v]

    # dynamic heuristic (user-provided)
    def h(self, n):
        return H[n]

    def a_star_algorithm(self, start_node, stop_node):
        open_list = set([start_node])
        closed_list = set([])
        g = {start_node: 0}
        parents = {start_node: start_node}

        while len(open_list) > 0:
            n = None
            for v in open_list:
                if n is None or g[v] + self.h(v) < g[n] + self.h(n):
                    n = v

            if n is None:
                print("Path does not exist!")
                return None

            if n == stop_node:
                reconst_path = []
                while parents[n] != n:
                    reconst_path.append(n)
                    n = parents[n]
                reconst_path.append(start_node)
                reconst_path.reverse()

                print("\n✅ Path found: {}".format(reconst_path))
                print("Total cost:", g[stop_node])
                return reconst_path

            for (m, weight) in self.get_neighbors(n):
                if m not in open_list and m not in closed_list:
                    open_list.add(m)
                    parents[m] = n
                    g[m] = g[n] + weight
                else:
                    if g[m] > g[n] + weight:
                        g[m] = g[n] + weight
                        parents[m] = n
                        if m in closed_list:
                            closed_list.remove(m)
                            open_list.add(m)

            open_list.remove(n)
            closed_list.add(n)

        print("Path does not exist!")
        return None


# ----------------- DYNAMIC SECTION -----------------
print("⭐ Dynamic A* Algorithm Input ⭐")

# Input all nodes
nodes = input("Enter all nodes (space-separated): ").split()

# Build adjacency list dynamically
adjacency_list = {}
for node in nodes:
    edges = []
    num_edges = int(input(f"\nEnter number of neighbors for node {node}: "))
    for _ in range(num_edges):
        neighbor = input(f"  Neighbor of {node}: ")
        weight = int(input(f"  Weight from {node} → {neighbor}: "))
        edges.append((neighbor, weight))
    adjacency_list[node] = edges

# Input heuristic values
H = {}
print("\nEnter heuristic values (h(n)) for each node:")
for node in nodes:
    H[node] = int(input(f"  h({node}) = "))

# Create graph
graph1 = Graph(adjacency_list)

# Input start and stop nodes
start = input("\nEnter start node: ")
stop = input("Enter goal node: ")

# Run A*
graph1.a_star_algorithm(start, stop)
```
